# Assignment 1: International Football Results Analysis

This notebook covers the full analysis exercise for the International Football Results dataset (1872–2024).

## Step 1: Load the CSV

First, we import the pandas library and load the results dataset into a DataFrame.

```python
import pandas as pd

try:
    df = pd.read_csv('../data/results.csv')
except FileNotFoundError:
    df = pd.read_csv('assignment1/data/results.csv')

df.head()
```

## Basic Exploration

### 1. How many matches are in the dataset?
We use the `.shape` attribute to find the total number of rows (matches).

```python
print(f"Total Matches: {df.shape[0]}")
```

### 2. What is the earliest and latest year in the data?
We convert the date column to datetime objects and use `.min()` and `.max()` to find the range.

```python
df['date'] = pd.to_datetime(df['date'])
print(f"Earliest Year: {df['date'].dt.year.min()}")
print(f"Latest Year: {df['date'].dt.year.max()}")
```

### 3. How many unique countries are there?
We combine the unique names from both home and away teams to find the total count of unique countries.

```python
unique_countries = set(df['home_team'].unique()).union(set(df['away_team'].unique()))
print(f"Unique Countries: {len(unique_countries)}")
```

### 4. Which team appears most frequently as home team?
We use `value_counts()` on the `home_team` column to find the most frequent team.

```python
print(df['home_team'].value_counts().head(1))
```

## Goals Analysis

First, we create a new column for total goals scored in each match.

```python
df["total_goals"] = df["home_score"] + df["away_score"]
```

### 5. What is the average number of goals per match?
We take the `.mean()` of the new `total_goals` column.

```python
print(f"Average Goals Per Match: {df['total_goals'].mean():.2f}")
```

### 6. What is the highest scoring match?
We find the maximum value and filter the dataframe to see the details.

```python
max_goals = df["total_goals"].max()
print(df[df["total_goals"] == max_goals][['date', 'home_team', 'away_team', 'home_score', 'away_score']])
```

### 7. Are more goals scored at home or away?
We compare the sums of `home_score` and `away_score`.

```python
home_sum = df["home_score"].sum()
away_sum = df["away_score"].sum()
print(f"Home Goals: {home_sum}, Away Goals: {away_sum}")
```

### 8. What is the most common total goals value?
We use the `.mode()` function to find the most frequent score.

```python
print(f"Most Common Total Goals: {df['total_goals'].mode()[0]}")
```

## Match Results

We create a helper function to determine the outcome of each match.

```python
def match_result(row):
    if row["home_score"] > row["away_score"]:
        return "Home Win"
    elif row["home_score"] < row["away_score"]:
        return "Away Win"
    else:
        return "Draw"

df["result"] = df.apply(match_result, axis=1)
```

### 9. What percentage of matches are home wins?
We use `value_counts(normalize=True)` to get the percentages.

```python
print(df["result"].value_counts(normalize=True) * 100)
```

### 10. Does home advantage exist?
We compare home win percentage vs away win percentage.

```python
res = df["result"].value_counts(normalize=True)
print("Home Win Advantage: ", res['Home Win'] > res['Away Win'])
```

### 11. Which country has the most wins historically?
We calculate wins by checking both home and away win scenarios.

```python
home_wins = df[df['result'] == 'Home Win']['home_team']
away_wins = df[df['result'] == 'Away Win']['away_team']
total_wins = pd.concat([home_wins, away_wins])
print(total_wins.value_counts().head(1))
```

## Visualization

### Histogram of goals
Visualizing the distribution of total goals.

```python
import matplotlib.pyplot as plt
df["total_goals"].hist(bins=15)
plt.title("Distribution of Goals Per Match")
plt.show()
```

### Bar chart of match outcomes
Comparing Home Wins, Away Wins, and Draws.

```python
df['result'].value_counts().plot(kind='bar')
plt.title("Match Outcomes")
plt.show()
```

### Top 10 teams by total wins
Displaying the top performing countries.

```python
total_wins.value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Teams by Total Wins")
plt.show()
```
