# Titanic: Feature Engineering (Assignment 2)

This notebook demonstrates the process of cleaning and engineering features for the Titanic dataset.

```python
import pandas as pd
import numpy as np

# Load data
try:
    train_df = pd.read_csv('../data/train.csv')
except FileNotFoundError:
    train_df = pd.read_csv('data/train.csv')

train_df.head()
```

## 1. Data Cleaning

Fill missing values for Age and Embarked.

```python
def clean_data(df):
    df['Age'] = df['Age'].fillna(df['Age'].median())
    df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
    df = df.drop(columns=['Cabin'])
    return df

cleaned_df = clean_data(train_df.copy())
cleaned_df.info()
```

## 2. Feature Engineering

Create FamilySize and encode Sex.

```python
def engineer_features(df):
    df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
    df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
    df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
    df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
    return df

engineered_df = engineer_features(cleaned_df.copy())
engineered_df.head()
```

## 3. Feature Selection

Drop non-predictive columns.

```python
def select_features(df):
    df = df.drop(columns=['PassengerId', 'Name', 'Ticket'])
    return df

final_df = select_features(engineered_df.copy())
print("Final columns:", final_df.columns.tolist())
```
